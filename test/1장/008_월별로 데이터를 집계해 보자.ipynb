```python
import pandas as pd
customer_master = pd.read_csv('customer_master.csv')
customer_master.head()
```

```python
item_master = pd.read_csv('item_master.csv')
item_master.head()
```

```python
transaction_1 = pd.read_csv('transaction_1.csv')
transaction_1.head()
```

```python
transaction_detail_1 = pd.read_csv('transaction_detail_1.csv')
transaction_detail_1.head()
```

### 002_데이터를 결합(유니언)해 보자

```python
transaction_2 = pd.read_csv('transaction_2.csv')
transaction = pd.concat([transaction_1, transaction_2], ignore_index=True) # pd.concat 으로 유니언 하는듯
transaction.head() # 유니언은 행 방향으로 늘리는것 (세로로 결합)
```

```python
# 데이터 개수 확인
print(len(transaction_1))
print(len(transaction_2))
print(len(transaction))

# transaction_1 과  transaction_2 를 합친 transaction 으로 5000 + 1785 = 6786
```

```python
# transaction_detail 유니언 하기
transaction_detail_2 = pd.read_csv('transaction_detail_2.csv')
transaction_detail = pd.concat([transaction_detail_1,transaction_detail_2],ignore_index=True)
# 유니언할때 ignore_index=True 가 필요한것으로 보인다 
transaction_detail.head()
```

### 003_매출 데이터끼리 결합(조인)해 보자

```python
join_data = pd.merge(transaction_detail, transaction[["transaction_id", "payment_date",
                                                     "customer_id"]], on="transaction_id", how="left")
# pd.merge 로 조인한다.
join_data.head()
```

```python
# 데이터 개수 알아보기
print(len(transaction_detail))
print(len(transaction))
print(len(join_data))
```

### 004_마스터데이터를 결합(조인)해 보자

```python
join_data = pd.merge(join_data, customer_master, on="customer_id", how="left") # pd.merge 를 이용해 조인
join_data = pd.merge(join_data, item_master, on="item_id", how="left")
join_data.head()
```

### 005_필요한 데이터 칼럼을 만들자

```python
join_data["price"] = join_data["quantity"] * join_data["item_price"] # quantity 와 item_price 의 곱이 price
join_data[["quantity", "item_price", "price"]].head() # quantity, item_price, price 나타내기
```

### 006_데이터를 검산하자

```python
# 각각의 price 총합 확인
print(join_data["price"].sum())
print(transaction["price"].sum())
```

```python
# 위와 같은것으로 True/False 로도 확인 가능하다.
join_data["price"].sum() == transaction["price"].sum()
```

### 007_각종 통계량을 파악하자

```python
join_data.isnull().sum() 

# 결손치의 개수를 출력하는 코드
# isnull() 을 이용하면 결손치가 True/False로 반환되기 때문에
# True의 개수를 칼럼마다 sum() 으로 계산한다.
# 이 데이터는 결손치가 없는 데이터이다.

# isnaull 함수는 결측값을 True로, 아닌값을 False로 반환한다.
# 이때 각각의 값에 대하여 결측값 여부를 검사하기 때문에
# sum 함수를 덧붙여 사용하면 각 칼럼마다 결측치가 몇개 있는지
# 확인할수가 있다.
# True가 1, False가 0

# 결측치란 값이 없는것을 의미한다.
```

```python
join_data.describe()

# 각종 통계량을 출력하는 코드
# describe() 를 이용하면 개수, 평균값, 표준편차, 최솟값, 사분위수, 중앙값, 최댓값 을 간단히 출력 가능
# describe() 는 숫자 데이터를 집계해준다.

```

```python
print(join_data["payment_date"].min())
print(join_data["payment_date"].max())

# 데이터의 기간이 2019년 2월 1일부터 2019년 7월 31일 까지임을 알수있다.
# min() 은 인수로 받은 자료형 내에서 최소값을 반환하는 함수
# max() 는 최대값을 반환
```

### 008_월별로 데이터를 집계해 보자

```python
join_data.dtypes

# 데이터 타입을 한번에 확인하려면 dtypes 을 사용하면된다.
# dtypes = 데이터타입
```

```python
join_data["payment_date"] = pd.to_datetime(join_data["payment_date"])
join_data["payment_month"] = join_data["payment_date"].dt.strftime("%Y%m")
join_data[["payment_date", "payment_month"]].head()

# dt 를 사용하면 년, 월의 추출이 가능하다.
# head 는 DataFrame 내의 처음 n줄의 데이터를 출력한다.
# 아무 입력이 없을경우 5줄을 기본으로 출력한다.

# strftime 사용시 적용 가능한 형식

# %y : 두 자리 수의 연도 ex) 19, 20, 21
# %Y : 네 자리 수의 연도 ex) 2019, 2020, 2021
# %m : 0을 채운 두 자리 수의 월 ex) 01, 02 ...  11 ,12
# %d : 0을 채운 두 자리 수의 일 ex) 01, 02 ...  30, 31
# %I : 0을 채운 12시간제의 시간 ex) 01, 02 … 12
# %H : 0을 채운 24시간제의 시간 ex) 00, 01 … 23
# %M : 0을 채운 두 자리 수의 분 ex) 00, 01 ... 58, 59
# %S : 0을 채운 두 자리 수의 초 ex) 00, 01 ... 58, 59
```

```python
join_data.groupby("payment_month").sum()["price"]

# groupby 는 집계연산(sum, count, mean, max, min...) 을 활용할때
# 쓰는 함수이다.

# groupby 는 집계하고싶은 칼럼(payment_month) 와 집계 방법(sum)
# 을 기술한다.
```
